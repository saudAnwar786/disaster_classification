```python
import cv2
import os
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# Paths
image_dir = '/kaggle/working/Comprehensive Disaster Dataset(CDD)/Human_Damage'
output_dir = '/kaggle/working/new_dir/Human_Damage'
os.makedirs(output_dir, exist_ok=True)

# Parameters
K = 36  # Number of clusters
image_size = (128, 128)  # Resize images for consistency


list_dir = ['sea','Human_Damage','human','Damaged_Infrastructure','Wild_Fire','Drought','Non_Damage_Buildings_Street','Non_Damage_Wildlife_Forest','Water_Disaster','Land_Slide','Urban_Fire']
```

```python

# Load and preprocess images
image_list = []
filenames = []

for file in os.listdir(image_dir):
        if file.endswith('.png'):
            try:
                img_path = os.path.join(image_dir, file)
                img = cv2.imread(img_path)
                img = cv2.resize(img, image_size)  # Resize for uniformity
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
                image_list.append(img.flatten())  # Flatten for clustering
                filenames.append(img_path)
            except:
                continue

# Convert image list to numpy array
image_array = np.array(image_list, dtype=np.float32)

# Apply K-means clustering
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(image_array)
labels = kmeans.labels_

# Select a representative image for each cluster
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, image_array)

for idx, image_idx in enumerate(closest):
    img_path = filenames[image_idx]
    img = cv2.imread(img_path)
    output_path = os.path.join(output_dir, f'cluster_{idx+1}.jpg')
    cv2.imwrite(output_path, img)  # Save the representative image

print(f"Saved {K} representative images to {output_dir}")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Dimensionality reduction for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(image_array)

# Plotting the results
plt.figure(figsize=(10, 7))
scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='tab20', marker='o')
plt.colorbar(scatter, label='Cluster')
plt.title('K-means Clustering of Human Damage Images')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()

print(f"Saved {K} representative images to {output_dir}")
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import numpy as np
import shutil

# Set source and destination paths
source_dir = "/kaggle/input/clustered-dataset/cluster_folder (1)"
destination_dir = "/kaggle/working/augmented_data"
target_count = 100  # Target number of images per subdirectory

# Ensure the destination directory exists
os.makedirs(destination_dir, exist_ok=True)

# Define ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Loop through each subdirectory in source
for sub_dir in os.listdir(source_dir):
    sub_dir_path = os.path.join(source_dir, sub_dir)
    
    if os.path.isdir(sub_dir_path):
        # Create corresponding subdirectory in the destination directory
        dest_sub_dir_path = os.path.join(destination_dir, sub_dir)
        os.makedirs(dest_sub_dir_path, exist_ok=True)
        
        # Get list of images in source subdirectory
        images = [img for img in os.listdir(sub_dir_path) if img.endswith(('png', 'jpg', 'jpeg'))]
        num_images = len(images)
        
        # Copy original images to the destination directory
        for img_name in images:
            shutil.copy(os.path.join(sub_dir_path, img_name), dest_sub_dir_path)
        
        # If fewer than target_count, augment images
        if num_images < target_count:
            for img_name in images:
                img_path = os.path.join(sub_dir_path, img_name)
                img = load_img(img_path)
                x = img_to_array(img)
                x = np.expand_dims(x, axis=0)
                
                # Augment the image and save it in the destination directory
                i = 0
                for batch in datagen.flow(x, batch_size=1, save_to_dir=dest_sub_dir_path, save_prefix='aug', save_format='jpg'):
                    i += 1
                    num_images += 1
                    if num_images >= target_count:
                        break
                        
            print(f"Augmented images in {sub_dir} to reach {target_count} images in {destination_dir}.")
```

```python
import zipfile
import os

def zip_folder(folder_path, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                # Add file to the zip file, using a relative path
                zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Example usage:
folder_path = '/kaggle/working/augmented_data'  # Path to the folder you want to zip
output_zip = 'augment_image.zip'  # Name of the output zip file

zip_folder(folder_path, output_zip)
print(f"Folder '{folder_path}' has been zipped as '{output_zip}'")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# create generator
datagen = ImageDataGenerator(rescale = 1./255,
                             shear_range = 0.2,
                             zoom_range = 0.2,
                             horizontal_flip = True,
                             vertical_flip=True,
                             rotation_range=30,
                             validation_split=0.15,
                             fill_mode='nearest'
                             )
```

```python
import os

# Directory path
directory_path = directory_train

# List to store file names
file_list = []

# Append file names to the list
for file_name in os.listdir(directory_path):
    file_list.append(file_name)

# Output the list of files
print(file_list)
```

```python
# prepare an iterators for each dataset
train_it = datagen.flow_from_directory( directory_train, 
                                       class_mode='categorical',
                                       classes=file_list,
                                       target_size=(224, 224),
                                       batch_size=32,
                                       shuffle=True,
                                       subset='training',
                                       seed=24)
```

```python
# Get the class indices
class_indices = train_it.class_indices
print("Class indices:", class_indices)

# Get the number of images for each class
class_counts = {class_name: 0 for class_name in class_indices}

# Calculate the number of images for each class
for class_name, index in class_indices.items():
    class_counts[class_name] = sum(train_it.labels == index)

# Print the number of images for each class
for class_name, count in class_counts.items():
    print(f"Number of images in '{class_name}' class: {count}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow.keras.layers as L
# import tensorflow_addons as tfa
import glob, random, os, warnings
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

print('TensorFlow Version ' + tf.__version__)

def seed_everything(seed = 0):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    os.environ['TF_DETERMINISTIC_OPS'] = '1'

seed_everything()

import warnings
warnings.filterwarnings("ignore")
```

```python
from vit_keras import vit
import tensorflow_addons as tfa

vit_model = vit.vit_b16(
        image_size = (224,224),
        activation = 'softmax',
        pretrained = True,
        include_top = False,
        pretrained_top = False,
        classes = 8)

model = tf.keras.Sequential([
        vit_model,
        tf.keras.layers.Flatten(),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(128, activation = tfa.activations.gelu),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(64, activation = tfa.activations.gelu),
        tf.keras.layers.Dense(32, activation = tfa.activations.gelu),
        tf.keras.layers.Dense(8, 'softmax')
    ],
    name = 'vision_transformer')

model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model,to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.01)
model.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
checkpoint_filepath = 'combined_model_d2_01.keras'
# Define the ModelCheckpoint callback
checkpoint = ModelCheckpoint(
    checkpoint_filepath,
    monitor='val_accuracy',   # Monitor validation accuracy
    save_best_only=True,      # Save only the best model
    mode='max',               # Save when validation accuracy improves
    verbose=1                 # Verbosity level
)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, precision_score, recall_score
import matplotlib.pyplot as plt

# Get predictions on validation data
y_pred = model.predict(val_it)
y_pred_classes = np.argmax(y_pred, axis=1)  # Predicted classes
y_true = val_it.classes  # True classes from validation data

# Calculate F1 Score, Precision, and Recall
f1 = f1_score(y_true, y_pred_classes, average='weighted')  # Weighted average for multi-class
precision = precision_score(y_true, y_pred_classes, average='weighted')  # Weighted average
recall = recall_score(y_true, y_pred_classes, average='weighted')  # Weighted average

# Print the metrics
print(f'F1 Score: {f1:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')

# Save the plot as a PNG file
plt.savefig('confusion_matrix.png')  # Save the figure as a PNG file

# Show the plot
plt.show()
```
